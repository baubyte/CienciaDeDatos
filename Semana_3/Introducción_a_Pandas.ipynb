## ¿Qué es Pandas?

Pandas es una librería especializada en:

- manipulación de datos
- análisis de datasets
- limpieza de información
- transformación de datos


Su estructura principal es el **DataFrame**, que funciona como una tabla similar a una hoja de cálculo o una tabla de base de datos.

Un DataFrame tiene:

filas → registros

columnas → variables

IMPORTACIÓN

```python

import pandas as pd

pd.set_option('display.float_format', '{:.2f}'.format)
```

## Crear un DataFrame
Un DataFrame puede construirse a partir de diferentes fuentes:

- diccionarios
- archivos CSV
- bases de datos
- APIs

En este ejemplo crearemos uno manualmente.

```python
datos = {
    "Nombre": ["Ana", "Juan", "María", "Pedro", "Laura"],
    "Edad": [25, 30, 22, 40, 28],
    "Ventas": [200, 150, 300, 250, 180],
    "Ciudad": ["Madrid", "Buenos Aires", "Madrid", "Bogotá", "Buenos Aires"]
}

df = pd.DataFrame(datos)

df
```

## SELECCIONAR COLUMNAS

```python
df["Edad"]
```

```python
#Seleccionar múltiples columnas:
df[["Nombre","Ventas"]]
```

## FILTRAR DATOS
En Ciencia de Datos es común seleccionar registros que cumplen ciertas condiciones.

```python
#devuelve solo las filas que cumplen la condición
df[df["Edad"] > 25]
```

```python
#Filtrar por texto:
df[df["Ciudad"] == "Madrid"]
```

## FILTROS MÚLTIPLES
Filtrado con múltiples condiciones

Utilizando operador & (AND):

```python
df[(df["Edad"] > 25) & (df["Ventas"] > 180)]
```

Utilizando operador | (OR):

```python
df[(df["Ciudad"] == "Madrid") | (df["Ciudad"] == "Bogotá")]
```

## OPERACIONES MATEMÁTICAS

```python
df["Edad"]
```

```python
# promedio de valores
df["Edad"].mean()
```

```python
# minimo
df["Edad"].min()
```

## INDEXACIÓN

### Indexación con LOC
 Permite seleccionar datos utilizando **etiquetas** (nombres de filas y columnas).

```python
#Seleccionar fila
df.loc[2]
```

```python
#Seleccionar fila y columna:
df.loc[2, "Nombre"]
```

```python
# Seleccionar varias columnas:
df.loc[:, ["Nombre","Ventas"]]
```

: significa todas las filas

###INDEXACIÓN CON ILOC
`iloc` permite seleccionar datos utilizando **posiciones numéricas**.

primer número → filas

segundo número → columnas

```python
#Seleccionar fila por posición:
df.iloc[2]
```

```python
#Seleccionar fila y columna:
df.iloc[2,1]
```

```python
#Seleccionar rango:
df.iloc[1:4]
```

```python
#Seleccionar filas y columnas específicas:
df.iloc[0:3, 1:3]
```

## CREAR NUEVAS COLUMNAS
En análisis de datos muchas veces necesitamos generar nuevas variables.

```python
df["Ventas_con_impuesto"] = df["Ventas"] * 1.21

df
```

## ORDENAR DATOS

```python
# Ordenar por ventas:
df.sort_values("Ventas")
```

```python
# Orden descendente:
df.sort_values("Ventas", ascending=False)
```

## AGRUPACIÓN

Una operación muy utilizada en Ciencia de Datos es agrupar información por categorías.

Calcula el promedio de ventas por ciudad.

```python
df.groupby("Ciudad")["Ventas"].mean()
```

Calcula la suma de ventas por ciudad

```python
df.groupby("Ciudad")["Ventas"].sum()
```
