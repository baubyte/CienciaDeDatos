## 1. Variables y Tipos de Datos

Una variable es un espacio en memoria donde almacenamos información.

Tipos básicos:
- int → números enteros
- float → números decimales
- str → texto
- bool → verdadero o falso

```python
# Ejemplos de variables

edad = 25
altura = 1.75
nombre = "Ana"
activo = True

print(type(edad))
print(type(altura))
print(type(nombre))
print(type(activo))
```

## 2. Estructuras de Datos - CONTENEDORES

Las más importantes:

- Listas → son colecciones ordenadas y modificables (mutables), ideales para conjuntos de datos dinámicos
- Tuplas → similares a listas pero inmutables (fijas), útiles para datos constantes
- Diccionarios → almacenan pares clave-valor
- Conjuntos → almacenan elementos únicos y desordenados

### Listas, list[]

```python
# Las listas se construyen usando []
colores = ["azul", "amarillo", "rojo", "verde"]
numeros = [1, 2, 3, 4, 5]

print(colores)
print(numeros)
```

```python
# las listas estan implementadas con el tipo `list`

type(colores)
```

```python
# las listas permiten tipos mezclados
[1, 2, "azul", 4, "verde", 6] # no es lo común
```

```python
# los contenedores también pueden guardar otros valores que sean contenedores

# una lista de tres listas de tres elementos
keypad = [
    [7,8,9],
    [4,5,6],
    [1,2,3]
]
print("keypad:", keypad)
```

### Diccionarios, dict {}

Representan un conjunto de pares claves-valor. Cada clave es única dentro del diccionario, y se utiliza para acceder a su valor asociado.

```python
#se representan con {clave: valor, ...}
codigos = {"BAJO": 20, "MEDIO": 50, "ALTO": 80 }

print(codigos)
```

```python
# los contenedores también pueden guardar otros valores que sean contenedores

# los datos de un automóvil
# Dic: auto; claves: Marca, modelo...lista: [1997, 2004, 2007]
auto = {
    "marca": "Ford",
    "modelo": "Focus",
    "anio": "1992",
    "revisiones": [1997, 2004, 2007]
}
print("auto:",auto)

print()
```

### Tuplas, tuples ( , )
Son colecciones ordenadas e inmutables de elementos, una vez creada, no se puede modificar sus elementos.

```python
# Se crean con parentesis y valores separados por comas,
error = (404, "not found")
print(error)
```

### Conjuntos: set { , , }
Los set son colecciones desordenadas de elementos únicos, no pueden contener elementos duplicados.


```python
# Se crean con {} y una lista de valores separados por comas

frutas = {'apple', 'orange', 'apple', 'pear', 'orange', 'banana'}
frutas
```

### Operaciones sobre contenedores

Largo, claves, valores

```python
# largo de un contendor (lista, tuple, set y dict)

colores = ["azul", "amarillo", "rojo", "verde"]
len(colores)
```

```python
codigos = {"BAJO": 20, "MEDIO": 50, "ALTO": 80 }
len(codigos)
```

```python
# las claves de un diccionario
codigos.keys()
```

```python
# los valores de un diccionario
codigos.values()
```

### Indexacion de contenedores
La indexación es la forma de acceder a un elemento o a un subconjunto de los elementos de una estructura.

```python
colores = ["azul", "amarillo", "rojo", "verde"]
codigos = {"BAJO": 20, "MEDIO": 50, "ALTO": 80 }
```

```python
#indexamos una lista
colores[2]
```

```python
# indexacion con valores negativos
colores[-1]
```

```python
#indexamos un dict
codigos["MEDIO"]
```

```python
# modificar una lista con sun indice
na = [1,2,3,4,5,6]
print("antes", na)

na[1] = 0
na[2] = 0

print("despues", na)
```
