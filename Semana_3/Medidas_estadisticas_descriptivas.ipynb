
## Tipos de datos
1.1 **Categóricos:** Nombran o identifican un grupo, pero no representan cantidades sobre las que tenga sentido hacer operaciones matemáticas.
No tienen magnitud numérica, aunque a veces se codifiquen con números.

**Categóricos nominales** : no responden a un orden especifico ej: colores

**Categóricos ordinales** : Son categorías que tienen un orden lógico, aunque no sean números en su forma original.

Solo pueden tomar un conjunto limitado de valores.
Existe un orden definido entre ellos.
Suelen representarse como etiquetas (strings).
Pueden codificarse numéricamente respetando ese orden.

Ejemplo:

Talles de ropa:

Etiquetas: XS, S, M, L, XL
Codificación posible: XS = 0, S = 1, M = 2, L = 3, XL = 4

**1.2 Numéricos continuos**

Pueden tomar infinitos valores dentro de un intervalo, incluyendo decimales (por ejemplo: altura, peso).

**1.3 Numéricos discretos**

Toman valores contables, generalmente enteros (por ejemplo: cantidad de hijos).

## 2) Medidas de tendencia central

### 2.1 Media
La **media** es el promedio aritmético. Sensible a valores extremos.

### 2.2 Mediana
La **mediana** es el valor central (50% por debajo y 50% por encima). Poco sensible a outliers.

### 2.3 Moda
La **moda** es el valor más frecuente (aplica a numéricos discretos y categóricos).


```python
#Generamos 1000 números continuos con media 50 y desviación estándar 10.
np.random.seed(0)  # Para reproducibilidad
data_cont = np.random.normal(loc=50, scale=10, size=1000)
```

```python
# Convertimos a DataFrame
df_cont = pd.DataFrame(data_cont, columns=['Cont Data'])
```

```python
df_cont
```

Vamos a visualizar utilizando un Histograma
A diferencia de un gráfico de barras donde se representan categorías individuales, en un histograma los datos numéricos se agrupan en rangos (o bins).

* Eje X: bins
* Eje Y: Frecuencia: la frecuencia con que los datos caen en cada rango.
Si en un intervalo (bin) hay más datos, la barra correspondiente será más alta.

Es útil para detectar patrones como:
- Simetría: Cuando los datos se distribuyen de forma pareja a ambos lados de la media, como en una distribución normal.
- Sesgo: Cuando los datos están más concentrados en un lado (sesgo positivo o negativo).
- Outliers: Valores que caen fuera del rango típico.


```python
# Graficamos un histograma para ver la distribución de los datos continuos
plt.figure(figsize=(8, 6))
plt.hist(df_cont['Cont Data'], bins=40, edgecolor='black', alpha=0.7)
plt.title('Distribución de Datos Continuos')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

### 2.1-Media (promedio)
La media es la suma de todos los valores dividida por el número de observaciones.La media es el promedio aritmético. Sensible a valores extremos.

```python
# Cálculo del promedio (media)
mean = df_cont['Cont Data'].mean()
mean
```

```python
# Graficar la media sobre el histograma
plt.figure(figsize=(8, 6))
plt.hist(df_cont['Cont Data'], bins=30, edgecolor='black', alpha=0.7)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plt.title('Distribución de Datos Continuos con Media')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

La línea roja indica el valor de la media, el cual puede estar cerca del centro si la distribución es simétrica.

### 2.2-Mediana
La mediana es el valor que divide a los datos en dos mitades iguales. Es útil cuando hay valores extremos (outliers) que podrían distorsionar la media.



```python
# Cálculo de la mediana
median = df_cont['Cont Data'].median()
```

```python
# Graficar la mediana sobre el histograma
plt.figure(figsize=(8, 6))
plt.hist(df_cont['Cont Data'], bins=30, edgecolor='black', alpha=0.7)
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')
plt.title('Distribución de Datos Continuos con Mediana')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

Explicación:
La línea verde muestra el valor de la mediana. Si la distribución es simétrica, la media y la mediana estarán cerca.

####Sesgo :

 El sesgo mide el grado de asimetría de una distribución en relación a su media. Si una distribución es simétrica, su sesgo será cero; si está sesgada hacia la derecha o hacia la izquierda, el sesgo será positivo o negativo, respectivamente.

```python
from scipy.stats import skew  # Para calcular el sesgo

# Cálculo del sesgo
sesgo = skew(df_cont['Cont Data'])
```

```python

# Graficar el histograma con el promedio y la mediana
plt.figure(figsize=(8, 6))
plt.hist(df_cont['Cont Data'], bins=30, edgecolor='black', alpha=0.7)
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Promedio: {mean:.2f}')
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Mediana: {median:.2f}')
plt.title(f'Distribución de Datos Continuos con Promedio y Mediana\nSesgo: {sesgo:.2f}')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

La función skew() de la librería scipy.stats se utiliza para calcular el sesgo o asimetría de una distribución de datos.

Resultados de skew():
* Sesgo positivo: Si el valor de skew() es positivo, significa que la distribución tiene una cola más larga hacia la derecha (hacia los valores altos). En este caso, la media suele estar por encima de la mediana.

* Sesgo negativo: Si el valor de skew() es negativo, significa que la distribución tiene una cola más larga hacia la izquierda (hacia los valores bajos). Aquí, la mediana es mayor que la media.


### 2.3 Moda
La **moda** es el valor más frecuente (aplica a numéricos discretos y categóricos).

```python
# Moda sobre una serie categórica
categorias = pd.Series(np.random.choice(['Rojo','Verde','Azul','Azul','Azul','Verde'], size=100))
moda_cat = categorias.mode()  # puede devolver más de una
moda_cat
```

```python
# Gráfico de barras simple para frecuencias categóricas
freqs = categorias.value_counts()
plt.figure()
freqs.plot(kind='bar')
plt.title("Frecuencia de categorías")
plt.xlabel("Categoría")
plt.ylabel("Frecuencia")
plt.show()
```

## 3) Medidas de dispersión

### 3.1- **Varianza**:
Mide cuánto se dispersan los datos en relación a la media.  
### 3.2- **Desviación estándar**:
La raíz cuadrada de la varianza, más fácil de interpretar porque tiene las mismas unidades que los datos.  
### 3.3- **Coeficiente de variación (CV)**:
`desviación_estándar / media` (compara dispersión **relativa** entre variables/grupos).
### 3.4- **Rango**: `máximo - mínimo`.

### 3.1-2 Varianza y Desviación Estándar

Varianza: Mide cuánto se dispersan los datos en relación a la media.
Desviación Estándar: La raíz cuadrada de la varianza, más fácil de interpretar porque tiene las mismas unidades que los datos.

```python
# Cálculo de la varianza y desviación estándar
var = df_cont['Cont Data'].var()
std = df_cont['Cont Data'].std()
```

```python
print(f'varianza: {var}')

print(f'Desviación Std: {std}')
```

```python
# Graficar un área que muestre la desviación estándar
plt.figure(figsize=(8, 6))
plt.hist(df_cont['Cont Data'], bins=30, edgecolor='black', alpha=0.7)
plt.axvline(mean - std, color='orange', linestyle='dashed', linewidth=2, label=f'Media - 1 Std : {mean - std:.2f}')
plt.axvline(mean + std, color='orange', linestyle='dashed', linewidth=2, label=f'Media + 1 Std : {mean + std:.2f}')
plt.title('Distribución de Datos Continuos con Desviación Estándar')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

Explicación:
Las líneas naranjas muestran el rango de una desviación estándar por encima y por debajo de la media. Esto nos ayuda a entender la dispersión de los datos.


Interpretación de si una desviación estándar es grande o pequeña depende del contexto y de los valores que estás analizando. Aquí hay algunas formas de interpretar este concepto:

Como la desciación estándar mide la dispersión: cuánto se alejan, en promedio, los datos de la media.

* Si es Pequeña → los valores están concentrados cerca de la media.

* Si es Grande → los valores están muy repartidos, lejos de la media.

* Regla empirica: 68-95-99.7 (si la distribución es normal): ayuda a interpretar la desviación estándar

~68% de los datos caen dentro de 1 desviación estándar (σ).

~95% dentro de 2σ.

~99.7% dentro de 3σ.

*Ejemplo:*
Supongamos que estás midiendo el tiempo que tardan los estudiantes en completar un examen y tienes dos grupos:

* Grupo A: tiempos promedio de 60 minutos, con una desviación estándar de 2 minutos.

* Grupo B: tiempos promedio de 60 minutos, pero con una desviación estándar de 10 minutos.

- En el Grupo A, la desviación estándar es pequeña (2 minutos), lo que indica que la mayoría de los estudiantes completan el examen cerca del promedio (alrededor de 60 minutos).

- En el Grupo B, la desviación estándar es grande (10 minutos), lo que sugiere que los tiempos varían mucho: algunos estudiantes terminan mucho antes y otros mucho después.

### 3.4-Rango y Rango Intercuartil (IQR)

Rango: Diferencia entre el valor máximo y el mínimo.

* Rango Intercuartil (IQR): Mide la dispersión de la mitad central de los datos, es decir, entre el primer cuartil (Q1) y el tercer cuartil (Q3).

```python
# Cálculo del rango y el rango intercuartil
rango_data = df_cont['Cont Data'].max() - df_cont['Cont Data'].min()
q1 = df_cont['Cont Data'].quantile(0.25)
q3 = df_cont['Cont Data'].quantile(0.75)
iqr = q3 - q1
```

```python
print(f'Rango: {rango_data}')

print(f'Rango intercuartil: {iqr}')
```

```python
# Graficar el rango intercuartil sobre el boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(df_cont['Cont Data'], vert=False)
plt.title('Boxplot de Datos Continuos (Rango Intercuartil)')
plt.show()
```

image.png

Explicación:

* La caja (IQR, Rango Intercuartil): Representa el 50% central de los datos. Se extiende desde Q1 (25%) hasta Q3 (75%).
Dentro de la caja está la “zona más común” de los valores.

* La línea dentro de la caja (Mediana, Q2): Marca el valor que está justo en el medio de la distribución.El 50% de los datos está a la izquierda y el otro 50% a la derecha.

* Los bigotes: Se extienden desde los bordes de la caja hasta el último valor que todavía se considera “normal”.Ese límite “normal” se calcula como 1.5 × IQR (Rango Intercuartil) hacia afuera desde Q1 y Q3.
Todo lo que esté dentro de los bigotes se considera parte de la variación normal de los datos.

* Los outliers (valores atípicos):Son los puntos que aparecen fuera de los bigotes. Indican valores muy alejados de la mayoría. En la práctica, suelen ser casos especiales que conviene revisar

Análisis de Outliers

Los outliers son valores que se encuentran fuera del rango típico de los datos, es decir, muy por debajo de Q1 o muy por encima de Q3. y pueden influir de manera importante en las medidas de tendencia central.

```python
# Identificación de outliers
outliers = df_cont[
    (df_cont['Cont Data'] < (q1 - 1.5 * iqr)) |   # valores menores a Q1 - 1.5*IQR
    (df_cont['Cont Data'] > (q3 + 1.5 * iqr))     # valores mayores a Q3 + 1.5*IQR
]

# Contar cuántos registros cumplen esas condiciones (cuántos outliers hay)
print(f'Número de outliers detectados: {len(outliers)}')
```
