## ¿QUÉ ES NUMPY?
 Es la biblioteca fundamental para el computo numerico en Python, que proporciona soporte para arreglos multidimensionales y funciones matemáticas optimizadas.
 En Ciencia de Datos es muy común trabajar con:
- vectores de datos
- matrices
- cálculos estadísticos
- operaciones matemáticas sobre grandes volúmenes de datos

NumPy permite realizar estas operaciones **de manera rápida y eficiente**.


- Ver doc en https://numpy.org/doc/stable/reference/arrays.ndarray.html
- Algunos tipos de numpy: https://wesmckinney.com/book/numpy-basics#tbl-table_array_dtypes

## IMPORTACIÓN
Por convención NumPy se importa como np.

```python
import numpy as np
```

## Listas

```python
# Array de NumPy
array = np.array([1,2,3,4])
```

Listas de Python vs Arrays de NumPy

```python
# Lista de Python
lista = [1,2,3,4]

print(lista)
print(array)
```

**Las listas** son flexibles y pueden tener distintos tipos de datos

**Los arrays NumPy** contienen un solo tipo de dato y están optimizados para cálculo.

#OPERACIONES MATEMÁTICAS
## Operaciones vectorizadas

Una gran ventaja de NumPy es que permite aplicar operaciones matemáticas sobre todos los elementos de un array **sin utilizar bucles**.

```python
datos = np.array([10,20,30,40])

print(datos + 5)
print(datos * 2)
```

## Arrays multidimensionales

En Ciencia de Datos muchas veces trabajamos con matrices.

Por ejemplo:

- filas → observaciones
- columnas → variables

```python
matriz = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print(matriz)
```

## PROPIEDADES DEL ARRAY

```python
print("Dimensiones:", matriz.ndim)
print("Forma:", matriz.shape)
print("Cantidad de elementos:", matriz.size)
```

Explicación:

ndim → número de dimensiones

shape → tamaño de la matriz

size → cantidad total de elementos

## Funciones estadísticas

NumPy incluye muchas funciones para análisis de datos.

```python
ventas = np.array([100, 120, 150, 90, 200])

print("Promedio:", ventas.mean())
print("Suma:", ventas.sum())
print("Valor máximo:", ventas.max())
print("Valor mínimo:", ventas.min())
```

## Operaciones


```python
arr1 = np.array([1, 2, 3, 4, 5, 6])

arr2 = np.array([10, 20, 30, 40, 50, 60])
```

```python
# operaciones con valores
arr1 + 10
```

```python
# operaciones con valores
arr1 * 2
```

```python
# operaciones con otras arrays

arr1 + arr2
```

```python
# podemos compararlos
arr1 > arr2
```

# INDEXACIÓN
## Acceso a elementos

```python
ventas = np.array([100,120,150,90,200])

print(ventas[0])
print(ventas[2])
```

```python
# los numpy arrays se pueden modificar con slices
ventas[1:2] = 0
ventas
```
