# Ejemplo de Codificacion

En el siguiente codigo se crea un dataframe utilizando pandas, a partir de un diccionario de listas y se realizan algunas acciones sobre el mismo.
**Observacion:** siempre es buena practica, cargar/generar los datos en una celda y realizar las tareas sobre el mismo en celdas separadas como se observa en el ejemplo

````python
import pandas as pd

# Crear un marco de datos de personas
data = {'Nombre': ['Juan', 'Laura', 'Sara', 'Antonio', 'Luisa'],
        'Edad': [25, 36, 40, 41, 29],
        'Género': ['M', 'F', 'F', 'X', 'F'],
        'Ciudad': ['Rosario', 'Rosario', 'Funes', 'Roldán', 'Rosario']}

df = pd.DataFrame(data)
df
````

````python
# Contar los valores únicos de la columna "Género"
conteo_genero = df['Género'].value_counts()
conteo_genero
````

````python
# Agrupar los datos por la columna "Ciudad"
agrupado_ciudad = df.groupby('Ciudad')['Nombre'].count()
agrupado_ciudad
````
