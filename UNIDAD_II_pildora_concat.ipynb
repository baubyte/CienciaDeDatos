#  UNIDAD II – Concatenación de datos con pandas

## ¿Cómo unimos datasets con la misma estructura?

Tenemos datos de ventas separados por meses.
Todos tienen la misma estructura.

¿Cómo los unimos en un solo dataset?

Importamos las dependencias

```python
import pandas as pd
```

```python
df_enero = pd.DataFrame({
    "producto": ["A", "B"],
    "ventas": [100, 200]
})

df_febrero = pd.DataFrame({
    "producto": ["A", "B"],
    "ventas": [150, 250]
})

df_enero, df_febrero
```

Cuando los datasets tienen la misma estructura, podemos **apilarlos**.

Para eso usamos `pd.concat()`

## Concatenar por filas (caso más común)

Agregamos filas (stack vertical)

```python
pd.concat([df_enero, df_febrero])
```

## Resetear índice

Evita índices duplicados

```python
pd.concat([df_enero, df_febrero], ignore_index=True)
```

## Concatenar por columnas

Agregamos columnas (lado a lado)

```python
pd.concat([df_enero, df_febrero], axis=1)
```

## Agregar claves (identificar origen)

Permite saber de qué dataset viene cada fila

```python
pd.concat([df_enero, df_febrero], keys=['enero','febrero'])
```

## Columnas diferentes

Si las columnas no coinciden, pandas completa con NaN

```python
df_marzo = pd.DataFrame({
    "producto": ["A", "B"],
    "ventas": [120, 220],
    "sucursal": ["X", "Y"]
})

pd.concat([df_enero, df_marzo], ignore_index=True)
```

## Controlar columnas con join

- outer (default): todas las columnas
- inner: solo las comunes

```python
pd.concat([df_enero, df_marzo], join='outer', ignore_index=True)
```

`concat` sirve para:

- juntar datasets iguales
- agregar filas o columnas

No combina información, solo la acumula
