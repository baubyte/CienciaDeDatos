# UNIDAD II - AGRUPACIÓN


Tenemos un dataset y queremos empezar a analizarlo.

Antes de aplicar técnicas más complejas, necesitamos entender los datos.

```python
import pandas as pd

df = pd.DataFrame({
    "categoria": ["A", "A", "B", "B", "C", "C", "A"],
    "sucursal": ["Norte", "Sur", "Norte", "Sur", "Norte", "Sur", "Norte"],
    "ventas": [100, 150, 200, 120, 300, 180, 90]
})

df
```

### Valores únicos

```python
df['categoria'].unique()
```

### Cantidad de valores distintos

```python
df['categoria'].nunique()
```

### Frecuencia de cada valor

```python
df['categoria'].value_counts()
```

### Ordenar resultados

```python
df['categoria'].value_counts().sort_values()
```

## Queremos ver patrones.

GroupBy permite agrupar datos y aplicar cálculos.

## Suma por categoría

```python
df.groupby('categoria')['ventas'].sum()
```

## Promedio por categoría

```python
df.groupby('categoria')['ventas'].mean()
```

## Cantidad de registros

```python
df.groupby('categoria')['ventas'].count()
```

## Agrupar por múltiples variables

```python
df.groupby(['categoria','sucursal'])['ventas'].sum()
```

## Múltiples métricas

```python
df.groupby('categoria')['ventas'].agg(['sum','mean','count'])
```
