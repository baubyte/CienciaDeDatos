# UNIDAD II-Píldora – Combinar datos con pandas (merge)

## ¿Cómo unimos información de distintas tablas?

Importamos las dependencias

```python
import pandas as pd
```

## EJEMPLO

Tenemos dos tablas:

- ventas
- productos

Queremos agregar información de productos a las ventas.

```python
ventas = pd.DataFrame({
    "producto_id": [1, 2, 3, 4],
    "ventas": [100, 200, 150, 300]
})

productos = pd.DataFrame({
    "producto_id": [1, 2, 3],
    "nombre": ["A", "B", "C"]
})

ventas, productos
```

Cuando queremos combinar información entre tablas usamos `merge`, permite combinar información entre tablas usando una clave en común

## Merge básico

```python
pd.merge(ventas, productos, on='producto_id')
```

Solo aparecen los registros que coinciden en ambas tablas

## Left join

Mantiene todas las ventas

```python
pd.merge(ventas, productos, on='producto_id', how='left')
```

## Right join

Mantiene todos los productos

```python
pd.merge(ventas, productos, on='producto_id', how='right')
```

¿Por qué en el right join no aparece el producto 4?

Porque ese producto no existe en la tabla de productos. Right join mantiene lo que existe en la tabla derecha.

Los joins no crean datos nuevos, Solo combinan los que ya existen

## Outer join

Mantiene todo de ambas tablas

```python
pd.merge(ventas, productos, on='producto_id', how='outer')
```

## Importancia de la clave

Si la clave está mal definida, el resultado también será incorrecto

## Conclusión

`merge` sirve para combinar información.

No apila datos (eso lo hacía concat)
Relaciona tablas mediante una clave
